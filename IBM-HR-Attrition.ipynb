# Feature Engineering

```python
#Create a new column named Total_Satisfaction
numeric_data['Total_Satisfaction']=(numeric_data['EnvironmentSatisfaction']+numeric_data['JobInvolvement']+numeric_data['JobSatisfaction']+numeric_data['RelationshipSatisfaction']+numeric_data['WorkLifeBalance'])/5
```

```python
# drop used column 
numeric_data.drop(['EnvironmentSatisfaction','JobInvolvement','JobSatisfaction','RelationshipSatisfaction','WorkLifeBalance'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
for column in cat_data.columns:
    cat_data[column]=LabelEncoder().fit_transform(cat_data[column])
```

```python
#Concat both numeric and cat data
data=pd.concat([numeric_data,cat_data],axis=1)
```

# Apply Algorithms

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X=data.drop('Attrition',axis=1)
y=data['Attrition']
```

```python
#Scaler the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
def algo(model):
    model.fit(X_train,y_train)
    pred=model.predict(X_test)
    print(f'accuracy_score= {accuracy_score(y_test,pred)}')
    print(f'classification_report=\n{classification_report(y_test,pred)}')
    print(confusion_matrix(y_test,pred))
```

```python
cls=LogisticRegression()
algo(cls)
```

```python
cls=RandomForestClassifier()
algo(cls)
```

```python
cls=XGBClassifier()
algo(cls)
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
kfold = KFold(n_splits=5, random_state=0,shuffle=True)
Accuracy = []
F1_score = []

classifiers = ['Logistic Regression',
              'Random Forest',
              'XGBoost']

models = [LogisticRegression(solver='liblinear'),
          RandomForestClassifier(n_estimators=100),
          XGBClassifier()
        ]

for i in models:
    model = i 
    accuracy =cross_val_score(model, X_train, y_train, cv=kfold,scoring='accuracy')
    f1_score =cross_val_score(model, X_train, y_train, cv=kfold,scoring='f1_macro')
    Accuracy.append(accuracy.mean())
    F1_score.append(f1_score.mean())
    
    
    
    
new_model_data_frame =pd.DataFrame({'Accuracy Mean': Accuracy,
                                   'f1 mean': F1_score},
                                   index=classifiers)

new_model_data_frame
```
